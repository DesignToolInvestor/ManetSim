```python
from sympy import Function, Symbol, diff, horner, pi, series, sin
from matplotlib import pyplot as plot
from LocUtil import LogGrid1
```

```python
nPlot = 100
```

```python
x = Symbol('x')
sinc_ = sin(pi*x) / (pi*x)
sinc_
```

```python
xL = LogGrid1(1e-8,10, nPlot)

y10L = [sinc_.evalf(subs={x:xV}, n=10) for xV in xL]
y20L = [sinc_.evalf(subs={x:xV}, n=20) for xV in xL]
err = [abs(y20 - y10) for (y10,y20) in zip(y10L,y20L)]

plot.loglog(xL,err)
```
