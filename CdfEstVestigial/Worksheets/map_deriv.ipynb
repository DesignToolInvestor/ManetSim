```python
from sympy import Symbol, Function, simplify, diff, evalf, log, sinc
from matplotlib import pyplot as plot
from LocUtil import Grid1
```

```python
nPlot = 100
```

```python
x = Symbol('x')
map = Function('phi')

map0 = log(x / (1 - x))
print(f'map0 = {map0}')

map1 = simplify(diff(map0, x))
print(f'map1 = {map1}')

map2 = simplify(diff(map0,x,x))
print(f'map2 = {map2}')
```

```python
xL = [(k + 0.5) / nPlot for k in range(nPlot)]
yL = [map0.evalf(subs={x:xV}) for xV in xL]

plot.plot(xL,yL)
```

```python
xL = [(k + 0.5) / nPlot for k in range(nPlot)]
yL = [map1.evalf(subs={x:xV}) for xV in xL]

plot.plot(xL,yL)
```

```python
xL = [(k + 0.5) / nPlot for k in range(nPlot)]
yL = [map2.evalf(subs={x:xV}) for xV in xL]

plot.plot(xL,yL)
```
