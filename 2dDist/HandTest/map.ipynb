```python
from matplotlib import pyplot as plot 
from sympy import Symbol, lambdify

from Map import LogRatio
from LocUtil import Grid1
```

```python
nPlot = 101
```

```python
x = Symbol('x')
z = Symbol('z')
```

```python
map = LogRatio(x,z)
mapF = lambdify(x, map.MapExp(), "math")
```

```python
xL = [(k + 0.5) / nPlot for k in range(nPlot)]
yL = [mapF(x) for x in xL]
plot.plot(xL,yL)
```

```python
map0 = LogRatio(x,z, (0,2))
invF0 = lambdify(z, map0.InvExp(), "math")

zL = Grid1(-7,7, nPlot)
x0L = [invF0(z) for z in zL]

plot.plot(zL,x0L)
```

```python
map1 = LogRatio(x,z, (0,2),2)
invF1 = lambdify(z, map1.InvExp(), "math")

x1L = [invF1(z) for z in zL]

plot.plot(zL,x0L)
plot.plot(zL,x1L)

plot.plot((-7,7), (1,1), '--', c="Black")
plot.plot((0,0), (0,2), '--', c='Black')
plot.plot((2,2), (0,2), '--', c='Black')
```

```python
map1 = LogRatio(x,z, (0,1),0,1)
invF1 = lambdify(z, map1.InvExp(), "math")

map2 = LogRatio(x,z, (0,1),0,2)
invF2 = lambdify(z, map2.InvExp(), "math")

x1L = [invF1(z) for z in zL]
x2L = [invF2(z) for z in zL]

plot.plot(zL,x1L)
plot.plot(zL,x2L)
```

## Derivatives

```python
map = LogRatio(x,z)
```

```python
map.DerivExp(1)
```

```python
map.DerivExp(2)
```

```python
map = LogRatio(x,z, (-1,1), 1, 2)
mapE = map.MapExp()
mapE
```

```python
deriv = mapE.diff(x,1).simplify()
```

```python
derivF = lambdify(x, deriv, "math")

xL = [2*(k + 0.5)/nPlot - 1 for k in range(nPlot)]
dL = [derivF(x) for x in xL]

plot.plot(xL, dL)
```
