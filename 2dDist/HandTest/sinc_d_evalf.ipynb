```python
from matplotlib import pyplot as plot
from sympy import Function, Integer, Symbol, pi, sin

from LocUtil import LogGrid1, Grid1

from Sinc import SincD
```

```python
nPlot = 101
```

```python
x = Symbol('x')
s = SincD(0, x)
```

```python
SincD(0.5, x)
```

```python
xV = 0.1
s.evalf(subs={x: xV})
```

```python
xL = Grid1(-0.02, 0.02, nPlot)
yL = [s.evalf(subs={x: xV}) for xV in xL]

plot.plot(xL,yL)
```

```python
xL = LogGrid1(1e-8, 1, nPlot)

y10L = [s.evalf(subs={x: xV}, n=10) for xV in xL]

alt = sin(pi*x) / (pi*x)
yAlt = [alt.evalf(subs={x: xV}, n=20) for xV in xL]

error = [abs(y10 - yAlt) for (y10,yAlt) in zip(y10L,yAlt)]
```

```python
plot.loglog(xL,error)
```
