# Inverse Square Bases Functions

```python
import DistExampB as db
import sympy as sp

from matplotlib import pyplot as plot
from numpy import sinc
from sympy import Symbol

from LocUtil import Grid1, LogGrid1, MinMax, SetSeed
from Map import LogRatio
```

```python
xSym = Symbol('x')
kSym = Symbol('k')
hSym = Symbol('h')
```

```python
nPlot = 101
```

```python
base = 1 / (1 + (xSym - kSym*hSym)**2)
```

```python
h = 1
```

```python
xGrid = Grid1(-6,6, nPlot)
yGrid = tuple(base.subs(hSym,h).subs(kSym,1).subs(xSym,x).evalf() for x in xGrid)

plot.plot(xGrid,yGrid, c='blue')
```
